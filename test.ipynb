```python
class System:
    def __init__(self, name: str, data_dict: dict):
        self.name = name #system name
        self.inner_dct = dict()
        
        for kk, vv in data_dict.items():
            if kk not in ["protein","ligand"]:
                raise ValueError(f"Incorrect data field for '{kk}': can only be protein or ligand")
            
            if type(vv) is not list and type(vv) is not str:
                raise TypeError(f"Incorrect data type for '{vv}' in '{self.name}, {kk}': can only be str or list")
            
            if type(vv) is list:
                self.inner_dct[kk] = vv
                setattr(self, kk, vv)
            
            elif type(vv) is str:
                self.inner_dct[kk] = [vv]
                setattr(self, kk, [vv])
    
    
    def __call__(self):
        if self.inner_dct or self.fname is None:
            return self.inner_dct

        raise ValueError('define the fname first!')
```

```python
def combinations(field:str,system:System): #predictor tmb aniria aqui
    """ 
    Returns the combinations of system sequences in a determinate field and its corresponding predictor input template
    """

    if field.lower() == "protein":
        return system.protein, chai_prot_fasta

    
    elif field.lower() == "ligand":
        return system.ligand, chai_lig_fasta
```

```python
def gen_subinputs(system:System): # Aquí també va predictor
    """ 
    Generates the input text that contain information about the sequences
    """
    list_subinputs = []
    for field in ["protein","ligand"]:
        try:
            data,txt = combinations(field, aa)
        except:
            continue
        list_subinputs.append(  "\n".join([txt.format(system=aa,seq=prot) for prot in data])   )
    
    return list_subinputs
```
